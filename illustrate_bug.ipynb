```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import numpy.fft as npf
```

```python
def get_data(fn, Nt, Ny, Nx):
    data = np.fromfile(fn)
    return (data[::2] + 1j*data[1::2]).reshape(Nt, Ny, Nx)
```

```python
# not working for 3d FFT
Nt, Ny, Nx = 2**4, 2**9, 2**13
Nt, Ny, Nx = 2**4, 2**13, 2**9
Nt, Ny, Nx = 2**8, 2**8, 2**12

# not working for both 3d and 2D+1D
Nt, Ny, Nx = 2**8, 2**10, 2**10
Nt, Ny, Nx = 2**8, 2**12, 2**8

# working
Nt, Ny, Nx = 2**4, 2**9, 2**9
Nt, Ny, Nx = 2**4, 2**9, 2**10
Nt, Ny, Nx = 2**4, 2**9, 2**11
Nt, Ny, Nx = 2**4, 2**9, 2**12
Nt, Ny, Nx = 2**4, 2**12, 2**9
Nt, Ny, Nx = 2**4, 2**10, 2**10
Nt, Ny, Nx = 2**8, 2**9, 2**9
Nt, Ny, Nx = 2**8, 2**9, 2**10
Nt, Ny, Nx = 2**8, 2**8, 2**11
Nt, Ny, Nx = 2**4, 2**12, 2**8

# eval
Nt, Ny, Nx = 2**8, 2**12, 2**8

Z3d = get_data('/tmp/z3d.bin', Nt, Ny, Nx)
Z21d = get_data('/tmp/z21d.bin', Nt, Ny, Nx)
Z3df = get_data('/tmp/z3d_fftw.bin', Nt, Ny, Nx)

fig, ((ax31, ax32, ax33),
      (ax211, ax212, ax213),
      (axf1, axf2, axf3)) = plt.subplots(ncols=3, nrows=3)
ax31.imshow(abs(npf.fftshift(Z3d[0])), aspect=Nx/Ny, interpolation='none', origin='lower')
ax32.imshow(abs(npf.fftshift(Z3d[:, 0])), aspect=Nx/Nt, interpolation='none', origin='lower')
ax33.imshow(abs(npf.fftshift(Z3d[:, :, 0])), aspect=Ny/Nt, interpolation='none', origin='lower')
ax211.imshow(abs(npf.fftshift(Z21d[0])), aspect=Nx/Ny, interpolation='none', origin='lower')
ax212.imshow(abs(npf.fftshift(Z21d[:, 0])), aspect=Nx/Nt, interpolation='none', origin='lower')
ax213.imshow(abs(npf.fftshift(Z21d[:, :, 0])), aspect=Ny/Nt, interpolation='none', origin='lower')
axf1.imshow(abs(npf.fftshift(Z3df[0])), aspect=Nx/Ny, interpolation='none', origin='lower')
axf2.imshow(abs(npf.fftshift(Z3df[:, 0])), aspect=Nx/Nt, interpolation='none', origin='lower')
axf3.imshow(abs(npf.fftshift(Z3df[:, :, 0])), aspect=Ny/Nt, interpolation='none', origin='lower')

fig.tight_layout()
```
